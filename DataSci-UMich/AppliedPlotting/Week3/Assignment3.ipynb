In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python

import matplotlib.colors as colors
import matplotlib.cm as cm
from scipy import stats
import ipywidgets as pywgt

```

```python
# store values of interest in dataframe indexed by Year
values_df = pd.DataFrame(df.mean(axis=1),columns=['Mean'])
values_df['Standard Deviation'] = df.std(axis=1)


# show bar plot of mean values for each year
import matplotlib.pyplot as plt

xpos = [1,2,3,4]
plt.figure(figsize=(8,5))
plt.bar(xpos, values_df['Mean'])
plt.show()
```

```python
# want 95% confidence interval, corresponding confidence coefficient value is 1.96
z = 1.96
# sample sizes
n = df.shape[1]

# now calculate the size of confidence intervals for each year of data
# standard deviation divided by square root of sample size * z
# ci = values_df['Standard Deviation'] / np.sqrt(n) * z
values_df['Confidence'] = z * values_df['Standard Deviation'] / np.sqrt(n)

values_df
```

```python
# now show plot with confidence intervalas as error bars and add correct labels
error = values_df['Confidence'] * 2

plt.figure(figsize=(8,5))
plt.bar(df.index,values_df['Mean'],yerr=error,capsize=10)
plt.xticks(df.index)
plt.xlabel('Year')
plt.ylabel('Mean Votes')
plt.show()
```

```python
# want user-selected y value to campare to confidence intervals
yvalue = int(input("Select y axis value: "))

# suggested value for showing color differences
# yvalue = 38000


values_df['Lower Bound'] = values_df['Mean'] - values_df['Confidence']
values_df['Upper Bound'] = values_df['Mean'] + values_df['Confidence']

values_df

# add coloring according to selection - blue white red
from matplotlib.colors import Normalize
from matplotlib.cm import get_cmap

colormap = get_cmap('bwr')
#normalized colormap according to size of confidence intervals
color_norm = Normalize(vmin= -z, vmax= z)

values_df['Color Value'] = color_norm( (values_df['Mean'] - yvalue) / values_df['Confidence'] )
colors = [colormap(x) for x in values_df['Color Value']]

# now plot with colormap relative to selected y value
plt.figure(figsize=(15,15))
plt.axhline(y=yvalue, linewidth=0.5, color="black")
plt.bar(df.index,values_df['Mean'],yerr=error,capsize=10,color=colors)
plt.xticks(df.index)

# create colorbar to indicate certainty levels
scalar_map = plt.cm.ScalarMappable(norm=color_norm, cmap=colormap)
scalar_map.set_array([])
colorbar = plt.colorbar(scalar_map, orientation="horizontal", drawedges=False)
colorbar.set_label('More Likely to be Lower <-------- --------> More Likely to be Higher', fontsize=12)

plt.title('Certainty Of Mean Value Relative To Y Value Selection Of {} '.format(yvalue), fontsize=20)
plt.ylabel('Number of Votes',fontsize=15)
plt.xlabel('Year of County Election', fontsize=15)
#plt.suptitle()

plt.show()
```
